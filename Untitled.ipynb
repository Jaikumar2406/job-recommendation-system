```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
cleanResume('hellp https://gmail and this one')
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(stop_words='english')

tv.fit(df['Resume'])
dataset = tv.transform(df['Resume'])
```

```python
#train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset , df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
#model_making
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.metrics import accuracy_score

cls = OneVsOneClassifier(KNeighborsClassifier())
cls.fit(X_train , y_train)
ypred = cls.predict(X_test)
print(accuracy_score(y_test , ypred))
```

```python
import re
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Text Preprocessing Function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Join tokens back into a single string
    return ' '.join(tokens)
```

```python
data['Experience'] = data['Experience'].apply(preprocess_text)
data['Qualifications'] = data['Qualifications'].apply(preprocess_text)
data['Job Title'] = data['Job Title'].apply(preprocess_text)
data['Job Description'] = data['Job Description'].apply(preprocess_text)
data['skills'] = data['skills'].apply(preprocess_text)
data['Responsibilities'] = data['Responsibilities'].apply(preprocess_text)
```

```python
data['combine_data'] = data['Experience'] + " " + data['Qualifications'] + " "+ data['Salary Range'] + " "+ data['location'] + " "+ data['Country'] + " "+ data['Work Type'] + " "+ data['Contact Person'] + " "+ data['Job Title'] + " "+ data['Job Description'] + " "+ data['Benefits'] + " "+ data['Responsibilities'] + " "+ data['Company']
```

```python
# Access first row using .iloc
print(data['combine_data'].iloc[0])  # Or use .loc if you know the row index
```

```python
vectorizer = TfidfVectorizer(max_features=500000)
```

```python
tfdf = vectorizer.fit_transform(data['combine_data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_jobs(input_title, top_n=10):
    input_title = preprocess_text(input_title)
    input_vector = vectorizer.transform([input_title])
    similarities = cosine_similarity(input_vector, tfdf).flatten()

    indices = similarities.argsort()[-top_n:][::-1]
    recommendations = data.iloc[indices]
    for rank, idx in enumerate(indices):
        row = data.iloc[idx]
        result = (
            f"Job {rank + 1}\n"
            f"Job Title: {row['Job Title']}\n"
            f"Company: {row['Company']}\n"
            f"Location: {row['location']}\n"
            f"Salary: {row['Salary Range']}k USD\n"
            f"Skills: {row['skills']}\n"
            f"Job Description: {row['Job Description']}\n"
            f"Qualifications: {row['Qualifications']}\n"
            f"Work Type: {row['Work Type']}\n"
            f"Responsibilities: {row['Responsibilities']}\n"
            f"Contact Person: {row['Contact Person']}\n"
            f"Experience: {row['Experience']}\n"
            f"Country: {row['Country']}\n"
        )
        print(result)
```
